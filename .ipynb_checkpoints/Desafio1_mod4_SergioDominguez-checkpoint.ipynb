Descripción
El conjunto de datos contiene un informe salarial fabricado en el que puedes realizar análisis
para tratar de estimar el salario basado en las condiciones dadas. Las columnas son las
siguientes:
● work_year: El año en que se pagó el salario.
● experience_level: El nivel de experiencia en el trabajo durante el año, con los
siguientes valores posibles: EN (nivel de entrada / junior), MI (nivel medio /
intermedio), SE (nivel senior / experto), EX (nivel ejecutivo / director).
● employment_type: El tipo de empleo para el puesto: PT (tiempo parcial), FT (tiempo
completo), CT (contrato), FL (trabajo freelance).
● job_title: El puesto de trabajo desempeñado durante el año.
● salary: El monto total del salario bruto pagado.
● salary_currency: La moneda en la que se pagó el salario, utilizando un código de
moneda ISO 4217.
● salary_in_usd: El salario en USD (tipo de cambio dividido por la tasa promedio de
USD para el año correspondiente a través de fxdata.foorilla.com).
● employee_residence: El país de residencia principal del empleado durante el año
laboral, utilizando un código de país ISO 3166.
● remote_ratio: La cantidad total de trabajo realizado de forma remota, con los
siguientes valores posibles: 0 (sin trabajo remoto, menos del 20%), 50 (parcialmente
remoto), 100 (totalmente remoto, más del 80%).
● company_location: El país de la oficina principal del empleador o sucursal
contratante, utilizando un código de país ISO 3166.
● company_size: El número promedio de personas que trabajaron para la empresa
durante el año: S (menos de 50 empleados, pequeña), M (50 a 250 empleados,
mediana), L (más de 250 empleados, grande).

```python
# importar librerias
import pandas as pd
import matplotlib as plt
```

```python
# cargar archivo con datos
df_Salarios = pd.read_csv('ds_salaries.csv')
#revisar
df_Salarios.shape
df_Salarios
```

1. El promedio general de los salarios, desviación estándar, quintiles y rango. (hint:
escoge la columna adecuada para comparar)

```python
# reviso todas la columnas numericas
df_Salarios.describe()
```

```python
# selecciono la columna "salary_in_usd", pues considero que es homogenea a la informacion de salarios
df_Salarios.describe()["salary_in_usd"]
```

```python
# los calcularemos en forma independiente
media = df_Salarios["salary_in_usd"].mean()  # Media
mediana = df_Salarios["salary_in_usd"].median()  # Mediana
moda = df_Salarios["salary_in_usd"].mode()
desviacion_estandar = df_Salarios["salary_in_usd"].std()  # Desviación estándar
variacion = df_Salarios["salary_in_usd"].var()  # variacion
minimo = df_Salarios["salary_in_usd"].min()  # Valor mínimo
maximo = df_Salarios["salary_in_usd"].max()  # Valor máximo
c1=df_Salarios["salary_in_usd"].quantile(0.25)
c3=df_Salarios["salary_in_usd"].quantile(0.75)
q1=df_Salarios["salary_in_usd"].quantile(0.20)
q2=df_Salarios["salary_in_usd"].quantile(0.40)
q3=df_Salarios["salary_in_usd"].quantile(0.60)
q4=df_Salarios["salary_in_usd"].quantile(0.80)
```

```python
print("Estadisticos significativos para columna salary_in_usd", sep='\n')
print("Medidas de tendencia central")
print(f"La media es: {media}")
print(f"La mediana es: {mediana}")
print(f"La moda es: {moda}", sep='\n')

print("indicadores de posicion")
print(f"El quintil  q1 es: {q1}")
print(f"El quintil  q2 es: {q2}")
print(f"El quintil  q3 es: {q3}")
print(f"El quintil  q4 es: {q4}", sep='\n')

print("Medidas de Dispersion")
print(f"La desviacion estandar es: {desviacion_estandar}")
print(f"El rango es: {maximo - minimo}")
print(f"El rango intercuartil es: {c3 - c1}")
print(f"El minimo es: {minimo}")
print(f"La maxima es: {maximo}")
print(f"El cuartil c1 es: {c1}")
print(f"La cuartil c3 es: {c3}")
```

2. Una comparación entre los salarios agrupados por 3 diferentes categorías a tu
elección. ¿Para cuál de ellas las medidas de tendencia central son más
representativas? ¿Para cuál son menos representativas? Explica.

```python
# ● experience_level: El nivel de experiencia en el trabajo durante el año, con los
# siguientes valores posibles: 
# EN (nivel de entrada / junior), 
# MI (nivel medio / intermedio), 
# SE (nivel senior / experto), 
# EX (nivel ejecutivo / director).
    
experiencia = pd.pivot_table(
    data=df_Salarios # DataFrame.
    , index='experience_level' # Columna(s) del DataFrame original que queremos como indice de la tabla resultante
    , values=['salary_in_usd'] # valores a lo que le aplicaremos la aggfunc.
    , aggfunc=['mean', 'median', 'std', 'var', 'min', 'max' ] # función a aplicar a la columna de valores
    , fill_value = 0  # tranforma Nan a cero
)
experiencia
```

analisis: selecione tres categorias de agrupamiento, a saber, experience_level, employment_type y remote_ratio.
y para responder la pregunta de representacion, encuentro que;

para el caso de agrupamiento de experience_level, las mayores diferencia entre el promedio y la mediana se 
produce en los niveles de "EN (nivel de entrada / junior)" y "SE (nivel senior / experto)", siendo la mediana mucho menor, 
deduciendo para estos casos una mayor diferencia salarial.

```python
# employment_type: El tipo de empleo para el puesto: 
# PT (tiempo parcial), 
# FT (tiempo completo), 
# CT (contrato), 
# FL (trabajo freelance).

TipoEmpleo = pd.pivot_table(
    data=df_Salarios # DataFrame.
    , index='employment_type' # Columna(s) del DataFrame original que queremos como indice de la tabla resultante
    , values=['salary_in_usd'] # valores a lo que le aplicaremos la aggfunc.
    , aggfunc=['mean', 'median', 'std', 'var', 'min', 'max' ] # función a aplicar a la columna de valores
    , fill_value = 0  # tranforma Nan a cero
)
TipoEmpleo
```

analisis: selecione tres categorias de agrupamiento, a saber, experience_level, employment_type y remote_ratio.
y para responder la pregunta de representacion, encuentro que;

para el caso de agrupamiento de employment_type, tanto "PT (tiempo parcial)" como "CT (contrato)" existe una brecha 
considerable entre el promedio y la media, lo que podemos deducir es que existe demasiada diferencia en el salario.

```python
# ● remote_ratio: La cantidad total de trabajo realizado de forma remota, con los
# siguientes valores posibles: 
# 0 (sin trabajo remoto, menos del 20%), 
# 50 (parcialmente remoto), 
# 100 (totalmente remoto, más del 80%).

Remoto = pd.pivot_table(
    data=df_Salarios # DataFrame.
    , index='remote_ratio' # Columna(s) del DataFrame original que queremos como indice de la tabla resultante
    , values=['salary_in_usd'] # valores a lo que le aplicaremos la aggfunc.
    , aggfunc=['mean', 'median', 'std', 'var', 'min', 'max' ] # función a aplicar a la columna de valores
    , fill_value = 0  # tranforma Nan a cero
)
Remoto
```

analisis: selecione tres categorias de agrupamiento, a saber, experience_level, employment_type y remote_ratio.
y para responder la pregunta de representacion, encuentro que;

para el caso de agrupamiento de remote_level los niveles de remuneracion entre el promedio y la mediana 
que estan mas cercanos es para "100 (totalmente remoto, más del 80%).", de esto deducimos que los sueldos en este grupo
se encuentran mas cercanos al promedio. y la mayor diferencia se produce en los "50 (parcialmente remoto)", esto quiere 
decir que existe mucha diferencia en el nivel de sueldo. 

3. ¿Qué cargos reciben mejores sueldos en las empresas con sede en Estados Unidos?

```python
# genero nuevo df con los registros de la sede US
df_Sede_Usa = df_Salarios[df_Salarios['company_location'] == 'US']
```

```python
# luego genero tabla pivote para columna job_tible y que genere el maximo de columna salary_in_usd
df_max = pd.pivot_table(
          data = df_Sede_Usa # DataFrame.
        , index=['job_title'] # Columna(s) del DataFrame original que queremos como indice de la tabla resultante
        , values=['salary_in_usd'] # valores a lo que le aplicaremos la aggfunc.
        , aggfunc=['max'] # función a aplicar a la columna de valores.
        , fill_value = 0  # tranforma Nan a cero
        )
```

```python
# ordeno de mayor a menor por colunma max salary_in_usd
df_max.sort_values(('max', 'salary_in_usd'), ascending=False)
```
